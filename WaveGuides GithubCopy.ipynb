```python
'''        
gives me model lambda_g vs frequency for each a value. 
We can fit these to the observed lambda_g vs freq using the chai squared method.
'''
c=3*10**8
chai=[]
freqs = np.array([7.5,7.75,8,8.25,8.5,9])

A = np.arange(2.2,2.3,0.001)
models = np.empty((len(A),len(freqs)))
for i in range (0,len(A)):
    a = A[i]*10**-2
    for j in range(0,6):
        f = freqs[j]*10**9
        t = c/f
        M = t/((1-(t/(2*a))**2)**.5)
        models[i,j] = M*100
'''Now, lets calculate the observed wavelengths vs frequency for open and shorted setups'''

g_o = []
g_s = []
err_o = []
err_s=[]

for i in range (0,12):
    N = vars()['n'+str(i)]
    
   
    
    if(i%2==0): 
        diff_o=0
        for j in range(1,len(N)):
            x =  N[j]-N[j-1]
            diff_o += x
            #print(diff_o)
        f = 2*diff_o/(len(N)-1)
        g_o.append(f)
        
        #error:
        rad_o = 0
        
        for h in range (1,len(N)):
            x =  N[h]-N[h-1]
            rad_o+= ((x-f)**2)/len(N)
        #print(rad_o**.5)
        err_o.append(rad_o**.5)
        
        
    if(i%2!=0):
        diff_s=0
        for h in range(1,len(N)):
            diff_s += N[h]-N[h-1]
        d = 2*diff_s/(len(N)-1)
        g_s.append(d)
       
        #error:
        rad_s = 0
       
        for w in range (1,len(N)):
            x =  N[w]-N[w-1]
            rad_s+= ((x-d)**2)/len(N)
        err_s.append(rad_s**.5)
```

```python
import DataAnalysis
chi_a = DataAnalysis.chi(g_s,g_o,models,2.2,2.3,0.001)
print('Best fit value of a = ', chi_a,'cm')
print('Very close to my value of 2.261 cm!')
```
